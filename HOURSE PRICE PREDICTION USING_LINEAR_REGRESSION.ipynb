```python
df['total_rooms']=np.log(df['total_rooms']+1)
df['total_bedrooms']=np.log(df['total_bedrooms']+1)
df['population']=np.log(df['population']+1)
df['households']=np.log(df['households']+1)
```

```python
# Count values type of ocean_proximity
df.ocean_proximity.value_counts()
```

```python
# Create dumies
df=df.join(pd.get_dummies(df.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

## Feature Engineering

```python
df.columns
```

```python
df['bedroom_ratio']=df['total_bedrooms']/df['total_rooms']
df['household_room']=df['total_bedrooms']/df['households']
```

```python
hm=df.corr()
plt.figure(figsize=(17,8))
sns.heatmap(hm,annot=True,cmap="coolwarm")
```

## Training Model

```python
from sklearn.model_selection import train_test_split
X=df.drop(['median_house_value'],axis=1)
y=df["median_house_value"]
```

```python
X.head(5)
```

```python
y.head(5)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
df.head(5)
```

```python
# Đánh giá hệ số chặn
print(lm.intercept_)
```

```python
lm.coef_
```

```python
X_train.columns
```

```python
cdf=pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf
```

# Predictions

```python
predictions=lm.predict(X_test)
```

```python
predictions# Đây là mức giá đựa đoán của căn nhà
```

```python
plt.figure(figsize=(12,7))
plt.scatter(y_test,predictions,color="red")# khi nó xếp thành một đường thẳng thì đây là một
                                           # dự đoán khá tốt
```

```python
plt.figure(figsize=(12,7))
sns.distplot((y_test-predictions),color="purple")# Phần còn lại được phân phối bình thường diều này chứng tỏ
# dự đoán đúng
# Nếu phân phối không bình thường thì có thể chúng ta đã bị sai ở một bước nào đó
```
